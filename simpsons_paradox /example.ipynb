```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('dataset.csv')
df.head()
```

```python
df.shape
```

```python
sns.scatterplot(x=df.iloc[:,0], y=df.iloc[:,1], alpha=0.5)
```

```python
sns.regplot(x=df.iloc[:,0], y=df.iloc[:,1], scatter_kws={'alpha':0.5}, line_kws={"linestyle":"--", "color": "red"})
```

```python
df.iloc[:,-1].unique()
```

```python
groups = df.groupby(df.columns[-1])

df1, df2, df3, df4 = [group for _, group in groups]
```

```python
sns.scatterplot(x=df1.iloc[:,0], y=df1.iloc[:,1], alpha=0.5)
sns.scatterplot(x=df2.iloc[:,0], y=df2.iloc[:,1], alpha=0.5)
sns.scatterplot(x=df3.iloc[:,0], y=df3.iloc[:,1], alpha=0.5)
sns.scatterplot(x=df4.iloc[:,0], y=df4.iloc[:,1], alpha=0.5)

plt.show()
```

```python
sns.regplot(x=df1.iloc[:,0], y=df1.iloc[:,1], scatter_kws={'alpha':0.5}, line_kws={"linestyle":"--", "color": "red"})
sns.regplot(x=df2.iloc[:,0], y=df2.iloc[:,1], scatter_kws={'alpha':0.5}, line_kws={"linestyle":"--", "color": "red"})
sns.regplot(x=df3.iloc[:,0], y=df3.iloc[:,1], scatter_kws={'alpha':0.5}, line_kws={"linestyle":"--", "color": "red"})
sns.regplot(x=df4.iloc[:,0], y=df4.iloc[:,1], scatter_kws={'alpha':0.5}, line_kws={"linestyle":"--", "color": "red"})


plt.show()
```
