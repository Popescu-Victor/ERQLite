```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('index_traits.csv')
df.drop(columns="Name", inplace=True)
df.columns = ["Index", "Starts", "End", "Duration", "Trait1", "Trait2", "Trait3", "Trait4", "Trait5", "Trait6", "Trait7"]
df.head()
```

```python
sns.heatmap(df.iloc[:,4:].corr(), annot=True)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(10,3))
sns.histplot(df['Duration'], ax = axes[0])
sns.histplot(df['Starts'], ax = axes[1], bins=15)
sns.histplot(df['End'], ax = axes[2], bins=15)
```

```python
fig, axe = plt.subplots(3,3, figsize=(10,10))

sns.regplot(data = df, x = "Duration", y = "Trait1", ax = axe[0,0], line_kws={"linestyle":"--", 'color':'red'})
sns.regplot(data = df, x = "Duration", y = "Trait2", ax = axe[0,1], line_kws={"linestyle":"--", 'color':'red'})
sns.regplot(data = df, x = "Duration", y = "Trait3", ax = axe[0,2], line_kws={"linestyle":"--", 'color':'red'})
sns.regplot(data = df, x = "Duration", y = "Trait4", ax = axe[1,0], line_kws={"linestyle":"--", 'color':'red'})
sns.regplot(data = df, x = "Duration", y = "Trait5", ax = axe[1,1], line_kws={"linestyle":"--", 'color':'red'})
sns.regplot(data = df, x = "Duration", y = "Trait6", ax = axe[1,2], line_kws={"linestyle":"--", 'color':'red'})
sns.regplot(data = df, x = "Duration", y = "Trait7", ax = axe[2,0], line_kws={"linestyle":"--", 'color':'red'})
```
